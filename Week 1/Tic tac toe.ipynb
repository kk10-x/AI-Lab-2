```python
#create board
board = [' ' for x in range(10)]
```

```python
#Make each move
def insertLetter(letter, pos):
    board[pos] = letter
```

```python
#check if current val is empty
def isEmptySpace(pos):
    return board[pos] == ' '
```

```python
#print board
def printBoard(board):
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
```

```python
#winning conditions
def hasWon(board, let):
    return (board[7] == let and board[8] == let and board[9] == let) or (board[6] == let and board[5] == let and board[4] == let) or (board[3] == let and board[2] == let and board[1] == let) or (board[1] == let and board[4] == let and board[7] == let) or (board[2] == let and board[5] == let and board[8] == let) or (board[3] == let and board[6] == let and board[9] == let) or (board[1] == let and board[5] == let and board[9] == let) or (board[3] == let and board[5] == let and board[7] == let)
```

```python
#player move making
def playerMove():
    run = True
    while run:
        move = input('Please select a position to a place an \'X\' (1-9)')
        try:
            move = int(move)
            if move > 0 and move < 10:
                if isEmptySpace(move):
                    run = False
                    insertLetter('X', move)
                else:
                    print('Sorry, this space is occupied')
            else:
                print('Please type a number within the range!')
        except:
            print('Please type a number!')
```

```python
#computer move making
def compMove():
    possibleMoves = [x for x, letter in enumerate(
        board) if letter == ' ' and x != 0]
    move = 0

    for let in ['O', 'X']:
        for i in possibleMoves:
            boardCopy = board[:]
            boardCopy[i] = let
            if hasWon(boardCopy, let):
                move = i
                return move

    cornersOpen = []
    for i in possibleMoves:
        if i in [1, 3, 7, 9]:
            cornersOpen.append(i)

    if len(cornersOpen) > 0:
        move = selectRandom(cornersOpen)
        return move

    if 5 in possibleMoves:
        move = 5
        return move

    edgesOpen = []
    for i in possibleMoves:
        if i in [2, 4, 6, 8]:
            edgesOpen.append(i)

    if len(edgesOpen) > 0:
        move = selectRandom(edgesOpen)
    return move
```

```python
#randomize move
def selectRandom(li):
    import random
    ln = len(li)
    r = random.randrange(0, ln)
    return li[r]
```

```python
#check if board is full
def isBoardFull(board):
    if board.count(' ') > 1:
        return False
    else:
        return True
```

```python
#driver code
def main():
    print('Welcome')
    printBoard(board)

    while not(isBoardFull(board)):
        if not(hasWon(board, 'O')):
            playerMove()
            printBoard(board)
        else:
            print('Sorry, O\'s won this game')
            break
        if not(hasWon(board, 'X')):
            move = compMove()
            if move == 0:
                print('Tie Game!')
            else:
                insertLetter('O', move)
                print('Computer placed an \'O\' in position', move, ':')
                printBoard(board)
        else:
            print('X\'s won this game. Congratulations!!!')
            break
    if isBoardFull(board):
        print('Tie Game!')
```

```python
main()
while True:
    k = input('Do you want to play again (y/n) ?').lower()
    if k in ['n', 'no']:
        break
    else:
        board = [' ' for x in range(10)]
        main()
```
